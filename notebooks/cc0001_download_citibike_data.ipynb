```python
import os
import pandas as pd
import time
import urllib
import zipfile

import warnings
warnings.filterwarnings('ignore')
```

```python
DATA_DIR = '../data'
```

```python
def peek(df):
    display(df.iloc[0:3, :])
    print(len(df))
```

```python
# TODO: Iterate over dates in loop programmatically.
dates = ['202110', '202109']

for date in dates:
    url = f'https://s3.amazonaws.com/tripdata/{date}-citibike-tripdata.csv.zip'
    filename_zip = f'{DATA_DIR}/citibike/zip/{date}-citibike-tripdata.csv.zip'

    print(f'Downloading: {url}')
    time_start = time.time()

    try:
        urllib.request.urlretrieve(url, filename_zip)
        time_end = time.time()
        print(f'Wrote file: {filename_zip} in {(time_end - time_start):0.2f}s')
    except urllib.error.HTTPError as error:
        print(f'HTTP error code {error.code} retrieving: {url}')

    with zipfile.ZipFile(filename_zip, 'r') as zip_file:
        zip_file.extractall(f'{DATA_DIR}/citibike/csv/')
```

```python
for date in dates:
    filename_csv = f'{DATA_DIR}/citibike/csv/{date}-citibike-tripdata.csv'
    df = pd.read_csv(filename_csv)
    peek(df)
```
