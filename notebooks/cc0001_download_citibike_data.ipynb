#### Description:
This notebook downloads all Citi Bike data from October 2013 - October 2021 and saves them as CSV files with consistent names.
 
#### Outputs:
```
data/citibike/csv/*-citibike-tripdata.csv
data/citibike/zip/*-citibike-tripdata*zip
```

````python
import os
import pandas as pd
import time
import urllib
import zipfile

import warnings
warnings.filterwarnings('ignore')
````

````python
DATA_DIR = '../data'
````

````python
def peek(df):
    display(df.iloc[0:3, :])
    print(len(df))
````

````python
YEAR_STRINGS = [str(y) for y in range(2014, 2021)]
MONTH_STRINGS = [f'{m:02d}' for m in range(1, 13)]

# Capture October 2013 - October 2021 to make year-over-year comparisons which
# include 2021 easier.
MONTH_2013_STRINGS = [f'{m:02d}' for m in range(10, 13)]
MONTH_2021_STRINGS = [f'{m:02d}' for m in range(1, 11)]

DATE_STRINGS = (
    [f'2013{m}' for m in MONTH_2013_STRINGS] +
    [f'{y}{m}' for y in YEAR_STRINGS for m in MONTH_STRINGS] +
    [f'2021{m}' for m in MONTH_2021_STRINGS]
)
print(DATE_STRINGS)

# Mark the date strings for which Citi Bike data is a .csv.zip  file.
CSV_ZIP_YEARS = set([str(y) for y in range(2017, 2022)])
CSV_ZIP_YEAR_DATE_STRINGS = [d for d in DATE_STRINGS if d[0:4] in CSV_ZIP_YEARS]
````

````python
for date in DATE_STRINGS:
    ext = 'csv.zip' if date in CSV_ZIP_YEAR_DATE_STRINGS else 'zip'
    url = f'https://s3.amazonaws.com/tripdata/{date}-citibike-tripdata.{ext}'
    filename_zip = f'{DATA_DIR}/citibike/zip/{date}-citibike-tripdata.{ext}'

    print(f'Downloading: {url}')
    time_start = time.time()

    try:
        urllib.request.urlretrieve(url, filename_zip)
        time_end = time.time()
        print(f'Wrote file: {filename_zip} in {(time_end - time_start):0.2f}s')
    except urllib.error.HTTPError as error:
        print(f'HTTP error code {error.code} retrieving: {url}')

    with zipfile.ZipFile(filename_zip, 'r') as zip_file:
        zip_file.extractall(f'{DATA_DIR}/citibike/csv/')
````

````python
# Some CSV files aren't named the same. Standardize the names.
RENAME_DATE_STRINGS = [
    '2013-10',
    '2013-11',
    '2013-12',
    '2014-01',
    '2014-02',
    '2014-03',
    '2014-04',
    '2014-05',
    '2014-06',
    '2014-07',
    '2014-08',
]
for date in RENAME_DATE_STRINGS:
    origin = f'{DATA_DIR}/citibike/csv/{date} - Citi Bike trip data.csv'
    destination = (f'{DATA_DIR}/citibike/csv/{date.replace("-", "")}' +
                   '-citibike-tripdata.csv')
    os.rename(origin, destination)
````

````python
for date in DATE_STRINGS:
    filename_csv = f'{DATA_DIR}/citibike/csv/{date}-citibike-tripdata.csv'
    df = pd.read_csv(filename_csv)
    print(filename_csv)
    peek(df)
````
